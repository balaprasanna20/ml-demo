```
import matplotlib.pyplot as plt
import pandas as pd
data={'x':[1,2,3,4,5,7,8,9],
'y':[5,4,6,7,1,5,8,9]}

df = pd.DataFrame(data)
multiply=df['x']*df['y']
df['z']=multiply

plt.plot(df['x'], df['y'], label='Line 1')
plt.plot(df['x'], df['z'], label='Line 2')
plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('GRAPH')
plt.show()
```

```
import matplotlib.pyplot as plt
import pandas as pd
data={'x':[1,2,3,4,5,7,8,9],
'y':[5,4,6,7,1,5,8,9]}

df = pd.DataFrame(data)
multiply=df['x']*df['y']
df['z']=multiply

plt.plot(df['x'], df['y'], label='Line 1',color="y")
plt.plot(df['x'], df['z'], label='Line 2',color="b")
plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('GRAPH')
plt.show()
```

```
import matplotlib.pyplot as plt
import pandas as pd
data={'x':[1,2,3,4,5,7,8,9],
'y':[5,4,6,7,1,5,8,9]}

df = pd.DataFrame(data)
multiply=df['x']*df['y']
df['z']=multiply

plt.plot(df['x'], df['y'], label='Line 1',color="y",ls=':')
plt.plot(df['x'], df['z'], label='Line 2',color="b",ls='dashdot')
plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('GRAPH')
plt.show()
```

```
import matplotlib.pyplot as plt
import pandas as pd
data={'x':[1,2,3,4,5,],'y':[7,4,5,8,9],'z':[1,2,3,4,5]}
df=pd.DataFrame(data)
fig,axs=plt.subplots(2,figsize=(8,5))
axs[0].bar(df['x'],df['y'])
axs[0].set_title('bar')
axs[0].set_xlabel('x')
axs[0].set_ylabel('y')
axs[1].bar(df['x'],df['z'])
axs[1].set_title('plot')
axs[1].set_xlabel('x')
axs[1].set_ylabel('y')
plt.show()
```
