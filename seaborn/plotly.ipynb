```
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[1,4,9,16,25]
plt.plot(x,y)
plt.xticks(x)
plt.yticks(y)
plt.title('graph')
plt.xlabel( 'X')
plt.ylabel('Y')
plt.show()

```

```
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_excel("/content/ml dataset.xlsx")
df
matrix=df.pivot_table(index='CASH OR GPAY',columns='OPERATION NAME',values='OPERATION COST')
sns.heatmap(matrix ,annot=True,cmap='coolwarm')
plt.title('Heatmap of Operation Cost by Payment Method and Operation')
plt.xlabel('Operation Name')
plt.ylabel('Payment Method')
plt.show()
```

```
import plotly.express as px
import pandas as pd
df=pd.read_excel('/content/ml dataset.xlsx')
fig=px.scatter(df,x='GENDER',y='PHONE.NO')
fig.update_layout(title='scatter')
fig.show()
```

```
import plotly.express as px
import pandas as pd
df=pd.read_excel('/content/ml dataset.xlsx')
fig=px.histogram(df,x='GENDER',y='PHONE.NO')
fig.update_layout(title='histogram')
fig.show()
```

```
import plotly.express as px
import pandas as pd
df=pd.read_excel('/content/ml dataset.xlsx')
fig=px.bar(df,x='GENDER',y='PHONE.NO')
fig.update_layout(title='bargraph')
fig.show()
```
